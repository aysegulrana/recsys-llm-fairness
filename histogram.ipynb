```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of languages
languages = ["english", "italian", "turkish", "persian"]

# Graph styling
sns.set(style="whitegrid")

# Plot the graphs for all four language files
for lang in languages:
    file_path = f"csv_files/combined_movies_{lang}.csv"

    df = pd.read_csv(file_path)

    print("FilePath: ", file_path)

    # Subgraphs
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    fig.suptitle(f"Language: {lang.capitalize()}", fontsize=16)

    # Genre histogram
    sns.countplot(data=df, x='genre', order=df['genre'].value_counts().index[:10], ax=axes[0], palette="viridis")
    axes[0].set_title('Distribution of Movie Genres')
    axes[0].set_xlabel('Count')
    axes[0].set_ylabel('Genre')
    axes[0].tick_params(axis='x', rotation=45)

    # Country histogram
    sns.countplot(data=df, x='country', order=df['country'].value_counts().index[:10], ax=axes[1], palette="plasma")
    axes[1].set_title('Top 10 Countries by Movie Count')
    axes[1].set_xlabel('Count')
    axes[1].set_ylabel('Country')
    axes[1].tick_params(axis='x', rotation=45)
```
